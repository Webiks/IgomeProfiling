```python
plt.figure(figsize=(20,10))

plt.subplot(1, 2, 1)
df = pd.read_csv(f'{data_prefix}.txt', delimiter='\t')
ax = sns.barplot(data=df, x="intersection_size", y="percent", hue="mAb", palette="muted")
ax.set_title('Dynamics of (unique) peptide presence across replicates')
ax.set_xlabel('Number of replicates sharing a peptide')
ax.set_ylabel('% out of total')
ax.set_yticks(range(0,101,10))
sns.set_style("ticks")

plt.subplot(1, 2, 2)
data_prefix = data_prefix.replace('peptides_intersection', 'motifs_intersection')
df = pd.read_csv(f'{data_prefix}.txt', delimiter='\t')
ax = sns.barplot(data=df, x="intersection_size", y="percent", hue="mAb", palette="muted")
ax.set_title('Dynamics of motif presence across replicates')
ax.set_xlabel('Number of replicates contribute to a motif')
ax.set_ylabel('% out of total')
ax.set_yticks(range(0,101,10))
sns.set_style("ticks")

plt.savefig(f'{os.path.split(data_prefix)[0]}/sharedness.png', dpi=600)
plt.show()
```
