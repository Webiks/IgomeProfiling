```python
peptides_data_prefix = '/Users/Oren/Dropbox/Projects/gershoni/IgOmeProfiling/sharedness/peptides_intersection'

f, (ax0, ax1) = plt.subplots(1, 2, figsize=(20,10))

df = pd.read_csv(f'{peptides_data_prefix}.txt', delimiter='\t')
# print(df)
sns.barplot(data=df, ax=ax0, x="intersection_size", y="percent", hue="mAb", palette="muted")
ax0.set_title('Presence dynamics of affinity-selected\npeptides across replicates')  # (unique)
ax0.set_xlabel('Number of replicates sharing a peptide')
ax0.set_ylabel('% out of total')
ax0.set_yticks(range(0,101,10))
sns.set_style("ticks")

motifs_data_prefix = peptides_data_prefix.replace('peptides_intersection', 'motifs_intersection')
df = pd.read_csv(f'{motifs_data_prefix}.txt', delimiter='\t')
# print(df)
sns.barplot(data=df, ax=ax1, x="intersection_size", y="percent", hue="mAb", palette="muted")
ax1.set_title('Presence dynamics of affinity-selected\npeptide motifs across replicates')
ax1.set_xlabel('Number of replicates contribute to a motif')
ax1.set_ylabel('% out of total')
ax1.set_yticks(range(0,101,10))
sns.set_style("ticks")

plt.rcParams.update({'font.size': 20})
plt.savefig(f'{os.path.split(motifs_data_prefix)[0]}/sharedness.png', bbox_inches="tight", dpi=1000)
plt.show()
```

```python
import pandas as pd
import dovpanda
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import numpy as np
import os
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
data_path = '/Users/Oren/Dropbox/Projects/Sweeps Project/Igome_paper/20_peps_vs_mAbs.csv'
df = pd.read_csv(data_path, index_col=0)
df
```

```python
cm = ['Greens', 'Reds']
f, (ax0, ax1) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [1, 3], 'wspace': 0.9})
sns.heatmap(np.log10(df.iloc[:,-1:]+1), annot=False, ax=ax0, cmap='inferno_r', fmt='.2f', cbar_kws={'label': 'Log10(Average Copy Number)', 'use_gridspec':False, 'location':'left', 'pad':0.2})
ax0.xaxis.set_ticks([])
ax0.yaxis.set_ticks([])
ax0.yaxis.label.set_visible(False)
sns.heatmap(df.iloc[:,:4], yticklabels=df.index, annot=True, ax=ax1, cmap='Greens', fmt='.2f', cbar_kws={'label': 'OD 650nm'})
ax1.yaxis.label.set_visible(False)
plt.hlines([7,11, 17, 20], *ax1.get_xlim())

plt.savefig(data_path.replace('csv', 'png'), bbox_inches="tight", dpi=1000)
plt.show()
plt.close()
```
